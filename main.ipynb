```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
drivers = pd.read_csv("data/drivers.csv")
```

```python
nationalities = drivers['nationality'].value_counts()
```

```python
total_drivers = len(drivers)
nationality_pcts = (nationalities / total_drivers) * 100

# Group categories with less than 2%
threshold = 2.0  # 2%
small_categories = nationality_pcts[nationality_pcts < threshold].sum()
nationality_pcts_grouped = nationality_pcts[nationality_pcts >= threshold]
nationality_pcts_grouped['Others'] = small_categories

plt.figure(figsize=(10, 8))
plt.pie(nationality_pcts_grouped.values, 
        labels=nationality_pcts_grouped.index, 
        autopct='%1.0f%%')
plt.title('F1 Drivers by Nationality')
plt.legend(nationality_pcts_grouped.index, 
          bbox_to_anchor=(1.05, 1), 
          loc='upper left')
plt.axis('equal')
plt.show()
```
